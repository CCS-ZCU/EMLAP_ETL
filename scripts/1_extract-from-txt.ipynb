## Read metadata from google sheet

```python
url = "https://docs.google.com/spreadsheets/d/1bkHHTYc86K2IuEXqfYfkDNt5LovtvCU3gvqHIbVio88/export?format=csv"
response = requests.get(url)
csv_content = io.BytesIO(response.content)
df = pd.read_csv(csv_content)
transposed_df = df.transpose()

# Use the first column from the original DataFrame as the column names for the transposed DataFrame
transposed_df.columns = df[df.columns[0]]

# Drop the first row from the transposed DataFrame
catalogue = transposed_df[1:]

# Display the DataFrame
catalogue.head(20)
```

```python
catalogue.to_pickle("../data/catalogue.pickle")
```
